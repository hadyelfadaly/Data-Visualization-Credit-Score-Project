# Requirment 1 
Load the dataset and show the first ten rows.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bank_loan = pd.read_csv(r'Dataset4.csv', low_memory=False)
```

```python
#Neither needed nor useful columns

bank_loan.drop(columns=['ID'], inplace=True)
bank_loan.drop(columns=['Customer_ID'], inplace=True)
bank_loan.drop(columns=['SSN'], inplace=True)
bank_loan.drop(columns=['Name'], inplace=True)
```

## Requirment 2.A
The type of distribution of the data.

```python
#lets pick Total_EMI_per_month as the example

col = 'Total_EMI_per_month'

plt.figure(figsize=(10, 5))

#plot the distribution
sns.histplot(bank_loan[col].dropna(), bins=50, kde=True)
plt.title(f'Right-Skewed Distribution of {col}')
plt.xlim(0, 5000) #zoom in to see the main part
plt.show()
```

## Requirment 2.B 
If there are any outliers in the data, show them.

```python
# define the columns you want to plot
columns = ['Num_Bank_Accounts', 'Num_Credit_Card', 'Interest_Rate', 'Num_Credit_Inquiries', 'Total_EMI_per_month']

plt.figure(figsize=(15, 12))

for i, col in enumerate(columns, 1):
    
    plt.subplot(2, 3, i)
    plt.boxplot(bank_loan[col].dropna(), vert=False)
    plt.xscale('log')  #log scale to compresses the huge values so we can see both the box and the outliers.
    plt.title(f'{col} (Log Scale)')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

## Requirement 2.C
Show the top three columns containing high variety in the number of categories in case of categorical features in the data; or the top three columns with the highest variance in case of all features were numeric.

```python
print(bank_loan.info())
```

```python
#define the lists of columns based on you the dat type we gonna change to
float_cols = ['Annual_Income', 'Monthly_Inhand_Salary', 'Changed_Credit_Limit', 'Outstanding_Debt', 'Payment_of_Min_Amount', 'Amount_invested_monthly', 
    'Monthly_Balance']
int_cols = ['Age', 'Num_of_Loan', 'Num_of_Delayed_Payment', 'Num_Credit_Inquiries']

#convert float columns
for col in float_cols:
    
    bank_loan[col] = pd.to_numeric(bank_loan[col], errors='coerce') #errors='coerce' turns text like "34_" or "Error" into NaN (blank)

#convert int columns
for col in int_cols:
    
    bank_loan[col] = pd.to_numeric(bank_loan[col], errors='coerce') #first clean it to numbers floats to handle the garbage text by the 'coerce'
    
    #convert to 'Int64' which allows integers with NaNs (as we will sill handle the nan values in the next preprocessing steps)
    bank_loan[col] = bank_loan[col].astype('Int64')

print(bank_loan.info())
```

```python
print("Top 3 Categorical Columns by Variety (Number of Unique Values)") #for categorical features

cat_cols = bank_loan.select_dtypes(include=['object', 'category']) #get only object/categorical columns

print(cat_cols.nunique().sort_values(ascending=False).head(3)) #count unique values and show top 3
print("\n" + "="*50 + "\n") #print new line
print("Top 3 Numeric Columns by Variance") #for numeric features

num_cols = bank_loan.select_dtypes(include=['number']) #get numeric columns

print(num_cols.var().sort_values(ascending=False).head(3)) #calculate variance and show top 3
```

## Requirement 2.D
If there are any correlations between variables.

```python
plt.figure(figsize=(16, 12)) #set the figure size (making it large enough to read the numbers)

corr_matrix = bank_loan.corr(numeric_only=True) #calculate the correlation matrix for numeric columns

#annot=True: Writes the number in each box
#fmt=".2f": Formats the numbers to 2 decimal places
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Variables', fontsize=16)
plt.show()
```
