```python
import matplotlib.pyplot as plt
import seaborn as sns

# We will focus on these numeric columns which likely have outliers
outlier_cols = ['Annual_Income', 'Age', 'Outstanding_Debt', 'Amount_invested_monthly']

# Set up the plotting area (2 rows, 2 columns)
plt.figure(figsize=(14, 10))

for i, col in enumerate(outlier_cols):
    plt.subplot(2, 2, i + 1)
    # Create the boxplot
    sns.boxplot(x=bank_loan[col], color='skyblue')
    plt.title(f'Boxplot of {col} (Outliers Detection)')
    plt.xlabel(col)

plt.tight_layout()
plt.show()

# --- Numeric Verification (Optional but Recommended) ---
# This calculates the exact count of outliers for 'Annual_Income' as an example
Q1 = bank_loan['Annual_Income'].quantile(0.25)
Q3 = bank_loan['Annual_Income'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = bank_loan[(bank_loan['Annual_Income'] < lower_bound) | (bank_loan['Annual_Income'] > upper_bound)]
print(f"Detected {len(outliers)} outliers in Annual_Income.")
print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
```
